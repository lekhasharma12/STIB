#Prompting Scores

```python
import pandas as pd
```

```python
zero_shot_df = pd.read_csv('../Results/Prompting/zero_shot_results.csv')
few_shot_df = pd.read_csv('../Results/Prompting/few_shot_cl_knn_results.csv')
instruction_df = pd.read_csv('../Results/Prompting/instruction_prompting_results.csv')
cot_df = pd.read_csv('../Results/Prompting/chain_of_thought_results_knn.csv')
```

```python
def calculate_avg_scores(df, score, prompting_strategy):
  df_m = df.groupby('Type').get_group('Metaphor')[score]
  df_i = df.groupby('Type').get_group('Idiom')[score]
  df_s = df.groupby('Type').get_group('Simile')[score]
  df_sa = df.groupby('Type').get_group('Sarcasm')[score]
  df_all = df[score]

  print(prompting_strategy +  " " + score + " Results:")
  print('Average metaphor: %f' %(sum(df_m)/len(df_m)))
  print('Average idiom: %f' %(sum(df_i)/len(df_i)))
  print('Average simile: %f' %(sum(df_s)/len(df_s)))
  print('Average sarcasm: %f' %(sum(df_sa)/len(df_sa)))
  print('Average total: %f' %(sum(df_all)/len(df_all)))
  print()
```

```python
calculate_avg_scores(zero_shot_df, "BERT", "Zero-shot")
calculate_avg_scores(zero_shot_df, "BLEURT", "Zero-shot")
calculate_avg_scores(zero_shot_df, "Score", "Zero-shot")
calculate_avg_scores(zero_shot_df, "Avg HE", "Zero-shot")
```

```python
calculate_avg_scores(few_shot_df, "BERT", "Few-shot")
calculate_avg_scores(few_shot_df, "BLEURT", "Few-shot")
calculate_avg_scores(few_shot_df, "Score", "Few-shot")
calculate_avg_scores(few_shot_df, "Avg HE", "Few-shot")
```

```python
calculate_avg_scores(instruction_df, "BERT", "Instruction prompting")
calculate_avg_scores(instruction_df, "BLEURT", "Instruction prompting")
calculate_avg_scores(instruction_df, "Score", "Instruction prompting")
calculate_avg_scores(instruction_df, "Avg HE", "Instruction prompting")
```

```python
calculate_avg_scores(cot_df, "BERT", "CoT")
calculate_avg_scores(cot_df, "BLEURT", "CoT")
calculate_avg_scores(cot_df, "Score", "CoT")
calculate_avg_scores(cot_df, "Avg HE", "CoT")
```
